### What This Cell Does
Cell 2 retrieves the current financial state

This is not optional housekeeping — it is

### Why Funds and Margins Matter Before Trading
Imagine arriving at a market without knowing
how much money is in your wallet. You might
commit to a purchase you cannot complete.
In algorithmic trading, the equivalent mistake
is placing an order when insufficient buying
power exists — which results in a rejected
order, a broken execution, and potentially
a cascade of errors downstream. Cell 2
prevents this by establishing the account's
financial state at the top of every session,
before any signal engine or order engine
is activated.

### The Three Numbers and What They Mean
The three values Cell 2 retrieves — cash,
buying power, and equity — are related but
distinct, and understanding the difference
matters as CLDSA moves toward live trading.

Cash is the actual settled money in the account
— funds that are unambiguously available and
not tied to any open position. In the paper
account this is $100,000.

Buying power is how much the account can
actually spend on new trades at this moment.
For Alpaca paper trading this is $200,000 —
double the cash balance — because Alpaca
offers 2x margin on paper accounts by default.
Margin means the broker is willing to lend
additional purchasing power against the
existing balance. In a live account, using
margin means using borrowed money, which
amplifies both gains and losses. This is
why CLDSA's Risk Parameter 1 caps individual
trade size at $1,500 regardless of available
buying power.

Equity is the total value of the account
if everything were liquidated right now —
cash plus the current market value of any
open positions. When no positions are held,
equity equals cash. When positions are open,
equity fluctuates with market prices, which
is precisely what Cell 7's portfolio tracker
monitors in real time.

a nested dictionary. The Indian equivalent

response and Upstox's nested dictionary
is one of the documented API quality
differences between the two brokers —
Alpaca returns data that can be accessed
directly, while Upstox requires an extra
extraction step to reach the same number.

```python
# Cell 3 — Historical Data via yfinance
# CLDSA — Cell 3
# Historical Data — yfinance
# 10-March-2026

from datetime import datetime
import pytz
import yfinance as yf
import pandas as pd

IST = pytz.timezone('Asia/Kolkata')
print(datetime.now(IST).strftime('%Y-%m-%d %H:%M:%S IST'))

# Symbols — yfinance format
# Alpaca uses US markets
SYMBOLS = {
    'AAPL':  'Apple Inc',
    'MSFT':  'Microsoft',
    'GOOGL': 'Alphabet'
}

print("\n===== HISTORICAL DATA — LAST 30 DAYS =====")

for ticker, name in SYMBOLS.items():
    try:
        data = yf.download(
            ticker,
            period='30d',
            interval='1d',
            progress=False
        )
        rows = len(data)
        latest_close = round(data['Close'].iloc[-1].values[0], 2)
        latest_date  = data.index[-1].strftime('%Y-%m-%d')
        print(f"\n{name} ({ticker}):")
        print(f"   Rows fetched:   {rows} days")
        print(f"   Latest close:   ${latest_close}")
        print(f"   Latest date:    {latest_date}")
    except Exception as e:
        print(f"❌ {ticker} failed: {e}")

print("\n==========================================")
print("Score!")
```

Without historical data, there is nothing
for the signal engine to reason about.

### What OHLCV Means and Why All Five Values Matter
OHLCV stands for Open, High, Low, Close,
and Volume. Each trading day produces exactly
these five numbers for every stock, and each
tells a different part of the day's story.

Open is the price at which the first trade
of the day occurred. High is the highest price
reached at any point during the day. Low is
the lowest. Close is the price of the last
trade before the market closed — and this is
the most important of the five for CLDSA's
purposes, because RSI and SMA are both
calculated on closing prices. The logic is
that the closing price represents the market's
final verdict on a stock's value for that day,
after all the intraday noise has settled.

Volume is how many shares changed hands
during the day. High volume on a price move
suggests the move is meaningful — many
participants agreed on that direction. Low
volume on a price move suggests it may be
fragile. CLDSA does not use volume in its
current signal logic, but it is retained
in the data because it becomes relevant
as the signal library grows.

### Why yfinance and Not the Broker API
Alpaca provides live price data beautifully —
that is what Cell 4 uses. But for historical
data going back 30, 60, or 90 days, yfinance
is the cleaner and more reliable source.
It pulls from Yahoo Finance's servers, requires
no API key or authentication, imposes no
rate limits for standard use, and works
identically regardless of which broker CLDSA
is connected to. This is what makes CLDSA
market-agnostic at the data layer — the
historical data pipeline works the same
whether the live trading happens through
Alpaca, Zerodha, or any future broker.

### The .values[0] Standard — Why It Exists
When yfinance returns data for a single ticker,
it wraps the Close column in a multi-level
index structure. This means that calling
data['Close'].iloc[-1] — which should return
a single number — actually returns a pandas
Series object instead. Trying to do arithmetic
on a Series when a single float is expected
causes silent errors or unexpected output.

The fix is .values[0], which unwraps the
Series and extracts the actual number.
This became CLDSA's documented standard
on 10-March-2026 after the issue was
encountered and diagnosed in Cell 3.
Every place in CLDSA where a single closing
price is extracted from yfinance data,
.values[0] is applied. This is one of
those standards that looks trivial in
the code but represents a genuine
diagnostic moment when first encountered.

Cell 3 and Cell 5A are a producer-consumer
pair. Cell 3 produces the DataFrame —
a table of historical prices. Cell 5A
consumes that DataFrame and derives RSI
and SMA values from it. The minimum
historical lookback required by the
indicators determines how many days
Cell 3 must fetch. SMA_50 needs at least
50 trading days to calculate meaningfully.
CLDSA uses 30 days for the current prototype,
which is sufficient for SMA_20 and RSI_14,
and will be extended to 90 days when SMA_50
becomes the primary signal component.

### What This Cell Does

### The Difference Between Historical and Live Data
Cell 3 and Cell 4 answer fundamentally
different questions. Cell 3 asks: what
has this stock been doing for the past
30 days? Cell 4 asks: what is this stock
worth right now, at this exact moment?

Historical data is stable — yesterday's
closing price will always be yesterday's
closing price, no matter when you ask
for it. Live data is a snapshot of a
moving target. The price Cell 4 returns
at 10:00 AM may be different by 10:01 AM.
This is why Cell 4B exists — to keep
refreshing that snapshot automatically —
but even Cell 4's single static fetch
is essential, because it gives the signal
engine in Cell 5B the current price to
compare against the historical indicators.

### What Bid and Ask Mean — and Why They Matter
Every actively traded stock at any given
moment has two prices, not one. The bid
price is the highest price any buyer in
the market is currently willing to pay.
The ask price is the lowest price any
seller is currently willing to accept.
The difference between these two numbers
is called the spread.

When CLDSA places a market buy order
in Cell 6, it buys at the ask price —
because that is what sellers are charging.
When it places a sell order, it sells at
the bid price — because that is what
buyers are offering. This means the spread
is an immediate, unavoidable cost that
CLDSA incurs the moment any trade is
executed. A stock with a wide spread —
say, a difference of Rs 5 on a Rs 100
stock — costs 5% just to enter and exit.
A tight spread — Rs 0.05 on a Rs 100
stock — is negligible.

Tracking the spread in Cell 4 is therefore
not academic curiosity. It is the beginning
of honest transaction cost accounting —
something that separates professional
algo systems from naive ones that look
only at closing prices and ignore the
cost of actually trading.

### The get_latest_trade Method
Alpaca provides several ways to fetch
current price data. CLDSA uses
get_latest_trade(), which returns the
most recent actual transaction that
occurred — a real trade between a real
buyer and seller. This is subtly different
from the mid-price, which is the mathematical
average of bid and ask and may not
correspond to any actual trade. Using
the latest trade price gives CLDSA a
grounded, real-world reference point
rather than a theoretical one.

### What This Cell Does
Cell 4B takes everything Cell 4 established —
a single live price fetch — and wraps it
in a continuous loop that repeats every
5 seconds. The result is a live price
monitor that keeps running until it is
deliberately stopped with Ctrl+C. This
is CLDSA's first taste of what a real
trading system feels like in operation —
a process that runs on its own, watching
the market, without requiring any human
input between refreshes.

### The while True Pattern — What It Means
The loop in Cell 4B is built using
Python's while True construct. To understand
why this works the way it does, it helps
to think about what "while True" literally
means to Python. A while loop keeps
running as long as its condition is True.
Since True is always True — it never
becomes False on its own — the loop
runs indefinitely. The only way to exit
is either through an explicit break
statement inside the loop, or through
an external interruption like Ctrl+C.

This is a very common pattern in
production systems that need to run
continuously — monitoring services,
data feeds, heartbeat checks. The loop
is not a mistake or a shortcut. It is
a deliberate design that says: keep
doing this until I tell you to stop.

### The 5-Second Interval — Why That Number
The time.sleep(5) call inside the loop
pauses execution for 5 seconds between
each price fetch. This number was chosen
as a balance between two competing needs.
On one side, a shorter interval — say
1 second — would give more frequent
price updates, which feels more "live."
On the other side, a shorter interval
means more API requests per minute,
which consumes rate limit allowance
faster and puts unnecessary load on
Alpaca's servers for a prototype system
that does not yet need millisecond
precision. 5 seconds is respectful of
the API, sufficient for a swing trading
prototype, and stable enough to run
for 19+ minutes without interruption
— as confirmed on 10-March-2026.

As CLDSA evolves toward Cell 4C and
WebSocket streaming, the polling interval
becomes irrelevant — because WebSocket
pushes data the moment it changes rather
than waiting to be asked. But for the
REST-based Cell 4B, 5 seconds represents
a well-reasoned engineering choice.

### The try/except Block — Why It Is Inside the Loop
Every fetch inside the loop is wrapped
in a try/except block. This is one of
CLDSA's core architectural standards,
and it deserves a careful explanation
because it is easy to add this pattern
mechanically without understanding why
it matters so much inside a loop
specifically.

When code runs once and fails, the
program simply stops and reports the
error. The damage is contained. But
when code runs inside a continuous
loop and fails without error handling,
the entire loop crashes — and all the
price data it was gathering stops
instantly, with no record of when or
why. By wrapping each iteration in
try/except, CLDSA ensures that a
single failed request — perhaps due
to a momentary network hiccup or
a brief API timeout — is logged and
skipped, while the loop continues
running. One missed data point is
acceptable. A crashed monitoring
system is not.

### The Kernel Blocking Behaviour — Important Workflow Note
One of the most practically significant
things Cell 4B reveals about Python and
Colab is the concept of kernel blocking.
While Cell 4B's while True loop is
running, the Colab kernel — the engine
that executes Python code — is entirely
occupied by that loop. No other cell
can run until the loop is stopped.

This means CLDSA has two operational
modes that cannot run simultaneously
in a single Colab session. Monitor mode
is when Cell 4B is running — watching
prices, but nothing else can execute.
Signal mode is when Cell 4B has been
stopped with Ctrl+C — and now Cells
5A, 5B, 6, and 7 can run, using the
most recently fetched price as their
input.

This is not a design flaw — it is a
constraint of the current single-kernel
architecture. Cell 4C, using WebSocket
streaming, will eventually resolve this
by handling the live data feed in a
background thread while the signal
engine runs in the foreground. For now,
the Ctrl+C discipline is CLDSA's
operational workaround, and it works.

### What a Technical Indicator Actually Is
It helps to understand what a technical
indicator is before learning what RSI and
SMA specifically mean. A technical indicator
is a mathematical transformation of price
data — a formula applied to a series of
historical prices that produces a new number
designed to reveal something about the
market's behaviour that is not immediately
obvious from looking at raw prices alone.
Raw prices tell you where the stock has been.
Indicators help interpret what that journey
might mean for where the stock is going.
No indicator predicts the future with
certainty — they are probabilistic tools
that identify conditions where certain
outcomes have historically been more likely.

### The Simple Moving Average — SMA
The SMA is the most conceptually
straightforward indicator. SMA_20 takes
the closing prices of the last 20 trading
days and calculates their arithmetic mean —
adds them all up and divides by 20. SMA_50
does the same across the last 50 trading
days. The result is a smoothed version of
price that filters out daily noise.

The reason two SMAs with different periods
are used together is that they reveal
different things. SMA_20 responds quickly
to recent price changes because it only
looks back 20 days. SMA_50 responds slowly
because it averages across 50 days and is
therefore less sensitive to short-term swings.
When the current price sits below both
moving averages — as it did for AAPL on

the stock may be in a downward trend or
at least below its medium-term average
cost, which is one condition that a
mean-reversion strategy would consider
potentially interesting.

### The Relative Strength Index — RSI
RSI is more sophisticated and worth
understanding carefully because it is
the trigger of CLDSA's current signal
logic. RSI was developed by J. Welles
Wilder and published in 1978. It measures
the speed and magnitude of recent price
changes to evaluate whether a stock is
being overbought or oversold by the market.

The calculation works by comparing the
average size of up-moves to the average
size of down-moves over the last 14 trading
days (RSI-14, which is the standard period).
If a stock has been closing higher more
often than lower, and by larger amounts,
the RSI will be high. If the reverse is
true — more and larger down-days — the
RSI will be low. The result is always
a number between 0 and 100.

The conventional interpretation is that
an RSI below 35 suggests the stock may
be oversold — meaning sellers may have
pushed the price down further than the
underlying fundamentals justify, and a
reversal upward becomes statistically
more probable. An RSI above 65 suggests
the stock may be overbought — buyers may
have driven the price higher than it can
sustainably hold. CLDSA's signal thresholds
of 35 and 65 are slightly more conservative
than the classical 30/70 levels, reflecting
the Risk Manager's preference for confirmed
signals over early entries.

world where its value changes every second
that the market is open. If AAPL's price
rises to $260.00 the next day, the position
is now worth $4.24 more than it cost.
If AAPL falls to $250.00, the position is
worth $5.76 less. This continuously updated
calculation of how much a position is worth
at current market prices — as opposed to
what it originally cost — is called marking
to market. The profit or loss that results
from this calculation is the unrealised PnL,
meaning it exists on paper but has not yet
been locked in by actually selling the position.

The word "unrealised" carries significant
weight. An unrealised gain is not the same
as money in the account. A stock that shows
a 10% unrealised gain can give back all of
that gain — and more — if the market turns.
This is why professional systems track
unrealised PnL carefully but do not treat
it as income until a position is closed
and the gain becomes realised. CLDSA's
Risk Parameter 3 — the 7% stop loss —
exists precisely to prevent an unrealised
loss from growing beyond an acceptable
threshold before it is converted into a
realised, contained loss.

### Why a Portfolio Tracker Is the Difference Between Flying Blind and Flying With Instruments
